```python
 # EXTRA Bubble Sort

def bubblesort(a,n):
  for i in range(0,n-1):
    for j in range(0,n-1-i):
      if a[j]>a[j+1]:
        temp=a[j]
        a[j]=a[j+1]
        a[j+1]=temp
  return a

a=list(map(int,input("Enter the numbers").split(",")))
n=len(a)
b=(bubblesort(a,n))
print(f"The sorted array is {b}")
```

```python
 # EXTRA Insertion Sort

def insertionsort(a,n):
  for i in range(1,n):
    j=i
    while(j>0 and a[j]<a[j-1]):
      temp=a[j]
      a[j]=a[j-1]
      a[j-1]=temp
      j-=1
  return a


a=list(map(int,input("Enter the numbers").split(",")))
n=len(a)
insertionsort(a,n)
```

```python
 # 25,26,27 Class, Object , Inheritance and Method overriding

class Vehicle:
  def __init__(self,brand,model,year):
    self.brand=brand
    self.model=model
    self.year=year
  def display_info(self):
    print(f"The brand ={self.brand}, Model ={self.model}, Year ={self.year}")

class Car(Vehicle):
    def __init__(self,brand,model,year,num_doors):
      super().__init__(brand,model,year)
      self.num_doors=num_doors
    def display_info(self):
      print(f"The brand ={self.brand}, Model ={self.model}, Year ={self.year}, number of doors ={self.num_doors}")
    def car_engine(self):
      print(f"Car {self.brand} has started")
class Bike(Vehicle):
    def __init__(self,brand,model,year,engine_capacity):
      super().__init__(brand,model,year)
      self.engine_capacity=engine_capacity
    def display_info(self):
      print(f"The brand ={self.brand}, Model ={self.model}, Year ={self.year}, engine capacity  ={self.engine_capacity }")
    def bike_engine(self):
      print(f"Bike {self.brand} has started")



vehicle=Vehicle('Mahindra','Truck',1998)
vehicle.display_info()
car=Car('BMW','C',1998,2)
car.display_info()
car.car_engine()
bike=Bike("Yamaha",'V4',2020,'500 HP')
bike.display_info()
bike.bike_engine()
```

```python
# 2. Diamon pattern using numbers

def increasing(i):
  return [i for i in range(1,i+1)]
def decreasing(i):
  return [i for i in range(i-1,0,-1)]

n=int(input("Enter the number of rows : "))
for i in range(1,n+1):
  inc=increasing(i)
  dec=decreasing(i)
  row=inc+dec
  print(" "*(n-i)+''.join(map(str,row)))
for i in range(n-1,0,-1):
  inc=increasing(i)
  dec=decreasing(i)
  row=inc+dec
  print(" "*(n-i)+''.join(map(str,row)))
```

```python
# 3. Spiral Matrix

n=int(input("Enter the size of matrix: "))
matrix=[[0 for _ in range(n)] for _ in range(n)]
top=0
bottom=n-1
left=0
right=n-1
num=1
while num<=n*n:
  for i in range(left,right+1):
    matrix[top][i]=num
    num=num+1
  top+=1
  for i in range(top,bottom+1):
    matrix[i][right]=num
    num+=1
  right-=1
  for i in range(right,left-1,-1):
    matrix[bottom][i]=num
    num+=1
  bottom-=1
  for i in range(bottom,top-1,-1):
    matrix[i][left]=num
    num+=1
  left+=1

for row in matrix:
    print(*row)
```

```python
# 4. Sieve of eratosthenes

def sieve_of_eratosthenes(n):
    primes = [True] * (n + 1)
    primes[0] = primes[1] = False  # 0 and 1 are not prime

    p = 2
    while p * p <= n:
        if primes[p]:
            for i in range(p * p, n + 1, p):
                primes[i] = False
        p += 1

    # Print all primes
    for i in range(2, n + 1):
        if primes[i]:
            print(i, end=' ')

# Example usage:
n = int(input("Enter upper limit: "))
sieve_of_eratosthenes(n)
```

```python
# 5. Permutations of a string

def permute(s, answer=""):
    if len(s) == 0:
        print(answer)
        return

    for i in range(len(s)):
        ch = s[i]
        left = s[:i]
        right = s[i+1:]
        rest = left + right
        permute(rest, answer + ch)

# Example usage
s = input("Enter a string: ")
permute(s)
```
